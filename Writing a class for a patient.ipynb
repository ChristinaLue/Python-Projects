```python
"""
In this project a patient - class with several class methods is defined. 
The class methods make it possible to update patient informations such as age and number of children and calculate the
patients insurance costs.
"""
class Patient:
    
    # initializer
    
    def __init__(self,name,age,sex,bmi,num_of_children,smoker):
        
        self.name = name
        self.age = age
        self.sex = sex # 0=female, 1=male
        self.bmi = bmi
        self.num_of_children = num_of_children
        self.smoker = smoker # 0=non-smoker, 1=smoker

    # class method that calculates a patients insurance cost
    
    def estimated_insurance_cost(self):
        
        estimated_cost = 250*self.age - 128*self.sex + 370*self.bmi + 425*self.num_of_children + 24000*self.smoker - 12500
        
        print(self.name + "'s estimated insurance cost is " + str(estimated_cost) + " dollars.")
        
    
    # class method that updates age of patient
    
    def update_age(self,new_age):
        
        self.age = new_age
        
        print(self.name + " is now " + str(self.age) + " years old.")
        
        # printing updated insurance cost (since it rises by age)
        self.estimated_insurance_cost()
        
    # class method that updates number of children
    
    def update_num_children(self,new_num_children):
        
        self.num_of_children = new_num_children
        
        if new_num_children == 1:
            
            print(self.name + " has " + str(self.num_of_children) + " child.")
            
        else:
            
            print(self.name + " has " + str(self.num_of_children) + " children.")
        
        self.estimated_insurance_cost()
        
        
    # class method that creates a dictionary of a patients data
    
    def patient_profile(self):
        
        patient_information = {"name": self.name, "age": self.age, "sex": self.sex, "bmi": self.bmi, "children": self.num_of_children,
                              "smoker": self.smoker}
        
        print(patient_information)


# Creating instance of the class Patient

patient1 = Patient("John Doe", 25, 1, 22.2, 0, 0)


# Checking if the class methods are working:

print(patient1)
# <__main__.Patient object at 0x0000016AD83374F0>

print(patient1.name)
# John Doe
print(patient1.age)
# 25
print(patient1.sex)
# 1 (male)
print(patient1.bmi)
# 22.2
print(patient1.num_of_children)
# 0
print(patient1.smoker)
# 0 (non-smoker)

print(patient1.estimated_insurance_cost())
# John Doe's estimated insurance cost is 1836.0 dollars.

print(patient1.update_age(26))
# John Doe is now 26 years old.
# John Doe's estimated insurance cost is 2086.0 dollars.

print(patient1.update_num_children(1))
# John Doe has 1 child.
# John Doe's estimated insurance cost is 2511.0 dollars.
```
